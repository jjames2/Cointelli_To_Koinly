```python
import pandas as pd
```

```python
final_cols = ['Date','Sent Amount','Sent Currency','Received Amount','Received Currency',
              'Fee Amount','Fee Currency','Net Worth Amount','Net Worth Currency','Label',
              'Description','TxHash']
col_dct = {'Out Quantity':'Sent Amount','Out Currency':'Sent Currency',
           'In Quantity':'Received Amount','In Currency':'Received Currency',
           'Fee Quantity 1':'Fee Amount','Fee Currency 1':'Fee Currency'}
def transformer(df):
    df.rename(columns=col_dct, inplace=True)
    df['datetime'] = pd.to_datetime(df['Date (UTC)'])
    df['Date'] = df['datetime'].dt.strftime('%Y-%m-%d %H:%m') + ' UTC'
    
    for col in final_cols:
        if col not in df.columns:
            df[col] = None
    df_final = df[final_cols].copy()
    return df_final
```
